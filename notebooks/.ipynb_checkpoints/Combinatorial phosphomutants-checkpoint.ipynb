# 3. Plot both together

```python
def plot_pair_combin(start, ax, min_x = 0, max_x = 0, top_label = False):
    phosphomut_bars = sog1_helpers.create_combo_df(start, PSv_conf2cA_STYA_vs_basic, PSv_conf2sA_STYA_vs_basic)
    phosphomim_bars = sog1_helpers.create_combo_df(start, PSv_conf2cD_vs_basic, PSv_conf2sD_vs_basic)
    both_bars = pd.merge(phosphomut_bars, phosphomim_bars, on = "combo_str", suffixes = ("_A", "_D"))
    both_bars["Activity_S3_1_A"] = both_bars["Activity_S3_1_A"] * -1
    
    plt.figure(figsize = (6, 2), dpi = 300)

    
    palette = sns.color_palette('rocket_r', n_colors = 4)
    palette_dict = {0: palette[0],
                    1: palette[1],
                    2: palette[2], 
                    3: palette[3]}
    
    # Plot the normal bar chart
    sns.barplot(y='combo_str', x='Activity_S3_1_D', data=both_bars, color='blue', edgecolor = 'none', hue = "count_D", 
                palette = palette, legend = False, zorder = 10, ax = ax)
    ax.axvline(both_bars[both_bars["combo_D"] == "WT"]["Activity_S3_1_D"].iloc[0], color = palette_dict[0], linestyle = "dotted", alpha = 1, zorder = 1, lw = 3)
    
    # Plot the reflected bar chart
    sns.barplot(y='combo_str', x='Activity_S3_1_A', data=both_bars, color='orange', edgecolor = 'none', hue = "count_A", palette = palette, legend = False, zorder = 10, ax = ax)
    ax.axvline(both_bars[both_bars["combo_D"] == "WT"]["Activity_S3_1_A"].iloc[0], color = palette_dict[0], linestyle = "dotted", alpha = 1, zorder = 1, lw = 3)

    if min_x == 0 and max_x == 0:
        min_x = min(both_bars["Activity_S3_1_A"])
        max_x = max(both_bars["Activity_S3_1_D"])
    ax.set_xlim(min_x -100, max_x + 100)
    
    sog1_helpers.plot_labels(start, palette_dict, "X", both_bars, ax, suffix = "_A", x = min_x - 550)
    
    ax.tick_params(axis='y', labelleft=False)
    ax.set_ylabel(str(start) + "-\n" + str(start+40), labelpad = 335, rotation = 0, ha = "center")
    ax.set_xlabel("Activity")
    
    ax.axvline(0, color = "black", lw = 2, zorder = 20)
    ax.axhline(0.5, color = "white", linestyle = "-", zorder = 0)
    ax.axhline(3.5, color = "white", linestyle = "-", zorder = 0)
    ax.axhline(6.5, color = "white", linestyle = "-", zorder = 0)

    if top_label:
        ax.text(x = min_x, y = -0.5, s = "Phosphomutant\n(X$\\rightarrow$A)", ha = "left", va = "bottom", fontsize = "small")
        ax.text(x = max_x, y = -0.5, s = "Phosphomimetic\n(X$\\rightarrow$D)", ha = "right", va = "bottom", fontsize = "small")
    # plt.text(x = min_x / 2, y = 10, s = "Phosphomutant\nActivity", ha = "center", va = "top", fontsize = "small")
    # plt.text(x = max_x / 2, y = 10, s = "Phosphomimetic\nActivity", ha = "center", va = "top", fontsize = "small")
    ax.set_xticklabels([int(abs(tick)) for tick in ax.get_xticks()])
    ax.set_yticks([])
    sns.despine(ax = ax, left = True)


    # Write N.D. where no data
    display(both_bars[both_bars["Activity_S3_1_A"].isna()])

    for i in both_bars[both_bars["Activity_S3_1_A"].isna()].index:
        ax.text(x = -1 * 75 * max_x / 5000, y = i, s= "N.D.", fontsize = "xx-small", va = "center", ha = "right")
```

```python
def find_bounds(starts):
    min_x = np.inf
    max_x = -np.inf
    
    for start in starts:
        phosphomut_bars = sog1_helpers.create_combo_df(start, PSv_conf2cA_STYA_vs_basic, PSv_conf2sA_STYA_vs_basic)
        phosphomim_bars = sog1_helpers.create_combo_df(start, PSv_conf2cD_vs_basic, PSv_conf2sD_vs_basic)
        both_bars = pd.merge(phosphomut_bars, phosphomim_bars, on = "combo_str", suffixes = ("_A", "_D"))
        both_bars["Activity_S3_1_A"] = both_bars["Activity_S3_1_A"] * -1
        
        min_x = min(min_x, min(both_bars["Activity_S3_1_A"]))
        max_x = max(max_x, max(both_bars["Activity_S3_1_D"]))

    return min_x, max_x
```

```python
sns.set_style('ticks')
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(5, 6), dpi = 300)
min_x, max_x = find_bounds([211, 221, 231])
plot_pair_combin(211, ax1, min_x, max_x, top_label = True)
plot_pair_combin(221, ax2, min_x, max_x)
plot_pair_combin(231, ax3, min_x, max_x)
```

```python
fig, (ax1) = plt.subplots(1, 1, sharex=True, figsize=(5, 2), dpi = 300)
plot_pair_combin(341, ax1, top_label = True)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(5, 4), dpi = 300)

starts = [401, 410]
min_x = np.inf
max_x = -np.inf

for start in starts:
    phosphomut_bars = sog1_helpers.create_combo_df(start, PSv_conf2cA_STYA_vs_basic, PSv_conf2sA_STYA_vs_basic)
    phosphomim_bars = sog1_helpers.create_combo_df(start, PSv_conf2cD_vs_basic, PSv_conf2sD_vs_basic)
    both_bars = pd.merge(phosphomut_bars, phosphomim_bars, on = "combo_str", suffixes = ("_A", "_D"))
    both_bars["Activity_S3_1_A"] = both_bars["Activity_S3_1_A"] * -1
    
    min_x = min(min_x, min(both_bars["Activity_S3_1_A"]))
    max_x = max(max_x, max(both_bars["Activity_S3_1_D"]))

plot_pair_combin(401, ax1, min_x, max_x, top_label = True)
plot_pair_combin(410, ax2, min_x, max_x, top_label = False);
```
