```python
HFB_FWYLtoA_s_vs_basic = sog1_helpers.add_var_positions(HFB_FWYLtoA_s, BasicArTh, activity_col, add_AAs = True)
HFB_FWYLtoA_s_vs_basic
```

```python
LtoA_s = HFB_FWYLtoA_s_vs_basic[HFB_FWYLtoA_s_vs_basic["AAs"] == "L"]
LtoA_s
```

```python
def single_AA_plot(AA, color, ax):
    LtoA_s = HFB_FWYLtoA_s_vs_basic[HFB_FWYLtoA_s_vs_basic["AAs"] == AA]
    #LtoA_s = LtoA_s[LtoA_s["mid"] < 115]
    plt.figure(figsize = (6, 2), dpi = 300)
    sns.lineplot(data = BasicArTh, x = "mid", y = activity_col, ax = ax)
    sog1_helpers.plot_all_tiles(LtoA_s, activity_col + "_var", ax=ax, color='gray', center = True)
    
    sns.scatterplot(data = LtoA_s, x = "var", y =  activity_col + "_var", color = color, marker = "D", alpha = 0.75, label = AA + "$\\rightarrow$A", s = 15, ax = ax)
    ax.legend(loc = "upper left")
    ax.set_ylabel("Activity")
    ax.set_xlabel("Tile Center")
    sns.despine()
```

```python
palette = sns.color_palette('colorblind')
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex = True, dpi = 300, figsize = (4, 5))

single_AA_plot("L", color = palette[1], ax = ax1)
sns.despine(ax = ax1)

single_AA_plot("F", color = palette[2], ax = ax2)
sns.despine(ax = ax2)

single_AA_plot("W", color = palette[3], ax = ax3)
sns.despine(ax = ax3)
single_AA_plot("Y", color = palette[4], ax = ax4)
sns.despine(ax = ax4)

ax1.set_ylabel("")
ax1.set_ylim(0, 5000)
ax2.set_ylabel("")
ax2.set_ylim(0, 5000)
ax3.set_ylabel("")
ax3.set_ylim(0, 5000)
ax4.set_ylabel("")
ax4.set_ylim(0, 5000)
fig.supylabel("Activity", x = -0.05)
```

```python
all_activities = sog1_helpers.return_activities("", pos_regex = False)
all_activities = all_activities.dropna(subset = [activity_col])
all_activities
```

```python
for AA in ["W", "F", "Y", "L", "D", "E"]:
    all_activities[AA] = all_activities["tile"].str.count(AA)
all_activities
```

```python
from matplotlib.ticker import MaxNLocator
```

```python
lightblue , darkblue, lightgreen, darkgreen, pink =  '#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99'

fig, axs = plt.subplots(1, 6, sharey = True, figsize = (8, 1.5), dpi = 300, gridspec_kw={'width_ratios': [1, 1, 1, 1, 2.5, 2.5]})
for AA, ax in zip(["W", "F", "Y", "L", "D", "E"], axs):
    #sns.boxplot(data = all_activities, x = AA, y = activity_col, ax = ax)
    color = lightblue if AA in ["D", "E"] else darkgreen

    sns.boxplot(x= AA,y=activity_col,data=all_activities,
                color = color, ax = ax, 
            fliersize = 0.1, 
            boxprops = {'edgecolor': "none"},
           flierprops = {'alpha' : 1, 'color' : "black", 'lw' : 0, 'markeredgewidth': 0},
           whiskerprops={'color': 'black', 'lw' : 0.5},  # whiskers set to black
            capprops={'color': 'black', 'lw' : 0.5},      # caps set to black
            medianprops={'color': 'black', 'lw' : 1})   # median line set to black) 
    
    sns.despine(ax = ax)
    ax.tick_params(axis='x', labelsize='xx-small', rotation=0)

plt.tight_layout(pad=0.1)

axs[0].set_ylabel("Activity")
```
